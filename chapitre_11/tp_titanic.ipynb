```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic = titanic[["survived","sex","age","pclass"]]
titanic
```

```python
titanic.dropna(axis=0,inplace=True)
titanic
```

```python
titanic['sex'] = titanic['sex'].replace(['male','female'],[0,1]).astype(int)
titanic
```

```python
y = titanic['survived']
X = titanic.drop('survived',axis=1)

from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd

classifier = KNeighborsClassifier()
classifier.fit(X,y)
classifier.score(X,y)

df_me = pd.DataFrame( np.array([0,10,1]).reshape(1,3), columns=X.columns)

classifier.predict(df_me)
```
